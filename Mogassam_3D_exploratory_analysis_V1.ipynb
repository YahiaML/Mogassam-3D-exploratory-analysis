```python
type_case_filter = df[df["Type"] == "Bridge 2"]

Error_range_values = list(type_case_filter["Error_range"].unique())

print((type_case_filter[type_case_filter["In_Normal_Case"] == True]["Type"].count() / type_case_filter["Type"].count()) * 100)
  
print((type_case_filter[type_case_filter["Error_range"] == "O"]["Error_range"].count() / type_case_filter["Error_range"].count()) * 100)
```

```python
#Choosing the type
for type_ in list(df["Type"].unique()):

  #Filtring
  type_filter = df[df["Type"] == type_]
  true_orient = type_filter[type_filter["In_Normal_Case"] == True]
  false_orient = type_filter[type_filter["In_Normal_Case"] == False]
  

  #Counting True/False orientation over all
  true_count = true_orient["In_Normal_Case"].count() 
  false_count = false_orient["In_Normal_Case"].count() 
  x_axis_mean = type_filter["X-axis"].mean()
  y_axis_mean = type_filter["Y-axis"].mean()
  z_axis_mean = type_filter["Z-axis"].mean()
  x_axis_sum = type_filter["X-axis"].sum()
  y_axis_sum = type_filter["Y-axis"].sum()
  z_axis_sum = type_filter["Z-axis"].sum()

  #Calculating the Percentage of True/False orientation over all
  true_present = (true_count / type_filter["In_Normal_Case"].count()) * 100
  false_present = (false_count / type_filter["In_Normal_Case"].count()) * 100

  #Counting True/False orientation over all
  true_obto = true_orient[true_orient["Error_range"] == "O.B.T.O"]["Error_range"].count()
  true_owob = true_orient[true_orient["Error_range"] == "O.W.O.B"]["Error_range"].count()
  true_normal = true_orient[true_orient["Error_range"] == "Normal"]["Error_range"].count()
  
  #Calculating the Percentage of True + Error range orientation over all
  true_obto_present = (true_obto / true_orient["Error_range"].count()) * 100
  true_owob_present = (true_owob / true_orient["Error_range"].count()) * 100
  true_normal_present = (true_normal / true_orient["Error_range"].count()) * 100

  #Counting True/False orientation over all
  false_low = false_orient[false_orient["Error_range"] == "Low"]["Error_range"].count()
  false_mediam = false_orient[false_orient["Error_range"] == "Mediam"]["Error_range"].count()
  false_high = false_orient[false_orient["Error_range"] == "High"]["Error_range"].count()
  
  #Calculating the Percentage of false + Error range orientation over all
  false_low_present = (false_low / false_orient["Error_range"].count()) * 100
  false_mediam_present = (false_mediam / false_orient["Error_range"].count()) * 100
  false_high_present = (false_high / false_orient["Error_range"].count()) * 100

  #Printing Results
  print("* "*20, type_,"* "*20)
  print(" ")
  print("- "*5, "#OVER ALL#", "- "*5)
  print("Percentage of True orientation =",round(true_present,2), "%", "(count = {})".format(true_count))
  print("Percentage of False orientation =", round(false_present,2), "%", "(count = {})".format(false_count))
  print("X-axis: mean =",x_axis_mean, " ,sum =",x_axis_sum)
  print("Y-axis: mean =",y_axis_mean, " ,sum =",y_axis_sum)
  print("Z-axis: mean =",z_axis_mean, " ,sum =",z_axis_sum)
  print(" ")
  print("- "*5, "#TRUE#", "- "*5)
  print("Percentage of Oriented better than the Original =", round(true_obto_present,2), "%", "(count = {})".format(true_obto))
  print("Percentage of Originally Wrong and oriented better =", round(true_owob_present,2),"%", "(count = {})".format(true_owob))
  print("Percentage of Normal orientation =", round(true_normal_present,2), "%", "(count = {})".format(true_normal))
  print(" ")
  print("- "*5, "#FALSE#", "- "*5)
  print("Percentage of Low error orientation =", round(false_low_present,2), "%", "(count = {})".format(false_low))
  print("Percentage of Mediam error orientation =", round(false_mediam_present,2),"%", "(count = {})".format(false_mediam))
  print("Percentage of High error orientation =", round(false_high_present,2), "%", "(count = {})".format(false_high))
  print(" ")
  print(" ")

```
