```python
import pandas as pd
```

```python
df =pd.read_excel("/content/Coordinates.xlsx")
```

```python
df.head()
```

```python
# whole data analysis

#Filtring
true_orient = df[df["In_Normal_Case"] == True]
false_orient = df[df["In_Normal_Case"] == False]
#Counting True/False orientation overall:
true_count = true_orient["In_Normal_Case"].count() 
false_count = false_orient["In_Normal_Case"].count() 

#Calculating the Percentage of True/False orientation overall:
true_present = (true_count / df["In_Normal_Case"].count()) * 100
false_present = (false_count / df["In_Normal_Case"].count()) *100

# Calculating overall means:
x_axis_mean = df["X-axis"].mean()
y_axis_mean = df["Y-axis"].mean()
z_axis_mean = df["Z-axis"].mean()

# overall Summation:
x_axis_sum = df["X-axis"].sum()
y_axis_sum = df["Y-axis"].sum()
z_axis_sum = df["Z-axis"].sum()

# Printing Results
print("* "*20, "All the data","* "*20)
print(" ")
print("Percentage of True orientation =",round(true_present,2), "%", "(count = {})".format(true_count))
print("Percentage of False orientation =", round(false_present,2), "%", "(count = {})".format(false_count))
print("X-axis: mean =",round(x_axis_mean,2), " ,sum =",round(x_axis_sum,2))
print("Y-axis: mean =",round(y_axis_mean,2), " ,sum =",round(y_axis_sum,2))
print("Z-axis: mean =",round(z_axis_mean,2), " ,sum =",round(z_axis_sum,2))
```

```python
# Each type analysis

# Formatting tools:
class format:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'

#Choosing the type
for type_ in list(df["Type"].unique()):

  #Filtring
  type_filter = df[df["Type"] == type_]
  true_orient = type_filter[type_filter["In_Normal_Case"] == True]
  false_orient = type_filter[type_filter["In_Normal_Case"] == False]
  
#--------------------------------------------------- Overall ---------------------------------------------------#

  #Counting True/False orientation overall:
  true_count = true_orient["In_Normal_Case"].count() 
  false_count = false_orient["In_Normal_Case"].count()  

  #Calculating the Percentage of True/False orientation overall:
  true_present = (true_count / type_filter["In_Normal_Case"].count()) * 100
  false_present = (false_count / type_filter["In_Normal_Case"].count()) *100
  
  # Calculating overall means:

  x_axis_mean = type_filter["X-axis"].mean()
  y_axis_mean = type_filter["Y-axis"].mean()
  z_axis_mean = type_filter["Z-axis"].mean()

  # overall Summation:
  x_axis_sum = type_filter["X-axis"].sum()
  y_axis_sum = type_filter["Y-axis"].sum()
  z_axis_sum = type_filter["Z-axis"].sum()

#--------------------------------------------------- True ---------------------------------------------------#

  # Calculating overall True means:

  true_x_axis_mean = true_orient["X-axis"].mean()
  true_y_axis_mean = true_orient["Y-axis"].mean()
  true_z_axis_mean = true_orient["Z-axis"].mean()

  # overall True Summation:
  true_x_axis_sum = true_orient["X-axis"].sum()
  true_y_axis_sum = true_orient["Y-axis"].sum()
  true_z_axis_sum = true_orient["Z-axis"].sum()

  # O.B.T.O axis means:
  true_obto_x_axis_mean = true_orient[true_orient["Error_range"] == "O.B.T.O"]["X-axis"].mean()
  true_obto_y_axis_mean = true_orient[true_orient["Error_range"] == "O.B.T.O"]["Y-axis"].mean()
  true_obto_z_axis_mean = true_orient[true_orient["Error_range"] == "O.B.T.O"]["Z-axis"].mean()

  # O.B.T.O axis Summation:
  true_obto_x_axis_sum = true_orient[true_orient["Error_range"] == "O.B.T.O"]["X-axis"].sum()
  true_obto_y_axis_sum = true_orient[true_orient["Error_range"] == "O.B.T.O"]["Y-axis"].sum()
  true_obto_z_axis_sum = true_orient[true_orient["Error_range"] == "O.B.T.O"]["Z-axis"].sum()

  #Counting True + Error orientation:
  true_obto = true_orient[true_orient["Error_range"] == "O.B.T.O"]["Error_range"].count()
  true_owob = true_orient[true_orient["Error_range"] == "O.W.O.B"]["Error_range"].count()
  true_normal = true_orient[true_orient["Error_range"] == "Normal"]["Error_range"].count()
  
  #Calculating the Percentage of True + Error range orientation:
  true_obto_present = (true_obto / true_orient["Error_range"].count()) * 100
  true_owob_present = (true_owob / true_orient["Error_range"].count()) * 100
  true_normal_present = (true_normal / true_orient["Error_range"].count()) * 100

#--------------------------------------------------- False ---------------------------------------------------#

  # Calculating overall True means:

  false_x_axis_mean = false_orient["X-axis"].mean()
  false_y_axis_mean = false_orient["Y-axis"].mean()
  false_z_axis_mean = false_orient["Z-axis"].mean()

  # overall True Summation:
  false_x_axis_sum = false_orient["X-axis"].sum()
  false_y_axis_sum = false_orient["Y-axis"].sum()
  false_z_axis_sum = false_orient["Z-axis"].sum()

  #Counting false + Error range orientation:
  false_low = false_orient[false_orient["Error_range"] == "Low"]["Error_range"].count()
  false_mediam = false_orient[false_orient["Error_range"] == "Mediam"]["Error_range"].count()
  false_high = false_orient[false_orient["Error_range"] == "High"]["Error_range"].count()
  
  #Calculating the Percentage of false + Error range orientation:
  false_low_present = (false_low / false_orient["Error_range"].count()) * 100
  false_mediam_present = (false_mediam / false_orient["Error_range"].count()) * 100
  false_high_present = (false_high / false_orient["Error_range"].count()) * 100

  #Printing Results
  print(format.BLUE +"* "*20, type_,"* "*20 + format.END)
  print(" ")
  print(format.BOLD +"- "*5, "#OVERALL#", "- "*5 + format.END)
  print("Percentage of True orientation =",round(true_present,2), "%", "(count = {})".format(true_count))
  print("Percentage of False orientation =", round(false_present,2), "%", "(count = {})".format(false_count))
  print("X-axis: mean =",round(x_axis_mean,2), " ,sum =",round(x_axis_sum,2))
  print("Y-axis: mean =",round(y_axis_mean,2), " ,sum =",round(y_axis_sum,2))
  print("Z-axis: mean =",round(z_axis_mean,2), " ,sum =",round(z_axis_sum,2))
  print(" ")
  print(format.BOLD + "- "*5, "#TRUE#", "- "*5 + format.END)
  print("X-axis: mean =",round(true_x_axis_mean,2), " ,sum =",round(true_x_axis_sum,2))
  print("Y-axis: mean =",round(true_y_axis_mean,2), " ,sum =",round(true_y_axis_sum,2))
  print("Z-axis: mean =",round(true_z_axis_mean,2), " ,sum =",round(true_z_axis_sum,2))
  print("Percentage of Oriented better than the Original 'O.B.T.O' =", round(true_obto_present,2), "%", "(count = {})".format(true_obto))
  print("Percentage of Originally Wrong and oriented better 'O.W.O.B' =", round(true_owob_present,2),"%", "(count = {})".format(true_owob))
  print("Percentage of Normal orientation =", round(true_normal_present,2), "%", "(count = {})".format(true_normal))
  print(" ")
  print(format.BOLD + "- "*5, "#FALSE#", "- "*5 + format.END)
  print("X-axis: mean =",round(false_x_axis_mean,2), " ,sum =",round(false_x_axis_sum,2))
  print("Y-axis: mean =",round(false_y_axis_mean,2), " ,sum =",round(false_y_axis_sum,2))
  print("Z-axis: mean =",round(false_z_axis_mean,2), " ,sum =",round(false_z_axis_sum,2))
  print("Percentage of Low error orientation =", round(false_low_present,2), "%", "(count = {})".format(false_low))
  print("Percentage of Mediam error orientation =", round(false_mediam_present,2),"%", "(count = {})".format(false_mediam))
  print("Percentage of High error orientation =", round(false_high_present,2), "%", "(count = {})".format(false_high))
  print(" ")
  print(" ")

```
